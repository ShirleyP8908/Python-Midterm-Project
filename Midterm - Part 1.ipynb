```python
#2. Load the data from day.csv as a Pandas data frame.
#Display its first 5 rows to confirm that the data is loaded correctly.


import pandas as pd
Bike_Data = pd.read_csv("day.csv")
```

```python
Bike_Data.head()
```

```python
#3. Display the shape of the data frame.
print("Shape:", Bike_Data.shape)
print("Column Names:")
print(Bike_Data.columns)
```

```python
#4. Display the data type of each column
Bike_Data.dtypes
```

```python
#5 The default data type for 'dteday' is object. Use pd.to_datetime to convert it to datetime type.
#After the conversion, display the data types again to confirm that the conversion is successful.

Bike_Data['dteday'] = pd.to_datetime(Bike_Data['dteday'])
Bike_Data.dtypes
```

```python
#6. Create a new column 'year' by extracting the year value from the datetime of each row.
from datetime import datetime

dteday = Bike_Data.loc[0,"dteday"]
print(dteday)


Bike_Data.loc[0,"Year"] = dteday.year
Bike_Data.head()
```

```python
import numpy as np

def extract_year(dteday):
    return dteday.year

extract_year(dteday)

Bike_Data['Year'] = Bike_Data['dteday'].apply(extract_year)

Bike_Data.head()
```

```python
#7 Remove the "casual" column and the "registered" column, as we are solely interested in the total number of bike rentals,
#which is given in the "cnt" column. Also remove the following columns: 'instant', 'yr', 'atemp'. 
#Display the first 5 rows again to confirm the removal of these columns.

Bike_Data.drop(['casual','registered', 'instant','yr','atemp'], axis= 'columns', inplace=True)
```

```python
Bike_Data.head()
```

```python
#8. Display the number of missing values from each column.
Bike_Data.isna().sum()
```

```python
# 5. The 'temp' feature: as temperature is a countinous feature, its relationship with the 'count' feature can be
#visualized as a scatter plot. Create a scatter plot with 'temp' and 'count'. Label the axis correctly.

import matplotlib.pyplot as plt
%matplotlib inline

plt.figure(figsize=(15,10))
plt.plot(Bike_Data['temp'], Bike_Data['cnt'], "b.")
plt.title("Temperature and Count relationship")
plt.xlabel("Temperature")
plt.ylabel("Count")
```

```python
#6. The 'windspeed' feature: create a scatter plot to show the relatioinship between 
# the number of bike rentals and the windspeed.

plt.figure(figsize=(15,10))
plt.plot(Bike_Data['windspeed'], Bike_Data['cnt'], "r.")
plt.title("Bike Rentals and Windspeed relationship")
plt.xlabel("Windspeed")
plt.ylabel("Count")
```
