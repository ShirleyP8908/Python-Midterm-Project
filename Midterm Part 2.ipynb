```python
#II. Create a Similarity Measure
#1. Calculate the standard deviation of each relavant feature.
#Store the reciprocals of the standard deviations in a weights list and print the list

# It combines the differences on each relevant feature: season, yr, mnth, holiday, weekday, workingday, weathersit, temp, hum, windspeed.
# Use absolute value of the differences to prevent cancelation between positive differences and negative differences.
# Use the reciprocal of feature standard deviations as the weights in the aggregation.

Standard_Dev = Bike_Data_2[['season', 'yr', 'mnth', 'holiday', 'weekday', 'workingday', 
                            'weathersit', 'temp', 'hum', 'windspeed']].std()
print(Standard_Dev)
```

```python
recipricol = np.reciprocal(Standard_Dev)
print("Reciprocals of each type:\n", recipricol)
```

```python
weights =  [0.900246, 1.998633, 0.289694, 5.982481, 0.498806, 2.149459,
            1.835218, 5.462959, 7.021037, 12.903580]
print(weights)
```

```python
#2. Create a similarity(test_row, data_row, weights) function that does the following:

# Calculate the absolute value of the difference between test_row 
# and data_row on each relevant feature.
# Calculate the sum of all differences, 
# each multiplied with its corresponding weight.
# Return the sum


import numpy as np
# The similarity measure should combine the differences from all features
def similarity(test_row, data_row, weights):
    
    # np.abs() calculates the absolute value
    diff1 = np.abs(test_row['season'] - data_row['season']) 
    diff2 = np.abs(test_row['yr'] - data_row['yr'])
    diff3 = np.abs(test_row['mnth'] - data_row['mnth'])
    diff4 = np.abs(test_row['holiday'] - data_row['holiday'])
    diff5 = np.abs(test_row['weekday'] - data_row['weekday'])
    diff6 = np.abs(test_row['workingday'] - data_row['workingday'])
    diff7 = np.abs(test_row['weathersit'] - data_row['weathersit'])
    diff8 = np.abs(test_row['temp'] - data_row['temp'])
    diff9 = np.abs(test_row['hum'] - data_row['hum'])
    diff10 = np.abs(test_row['windspeed'] - data_row['windspeed'])
    
    print("Differences:\n", diff1, diff2, diff3, diff4, diff5, diff6, diff7, diff8, diff9,diff10)

    similarity = weights[0] * diff1 + weights[1] * diff2 \
                + weights[2] * diff3 + weights[3] * diff4 \
                + weights[4] * diff5 + weights[5] * diff6 \
                + weights[6] * diff7 + weights[7] * diff8 \
                + weights[8] * diff9 + weights[9] * diff10
    return similarity
```

```python
# 3. Test the similarity function by 
# calculating the similarity between the first row 
# in the test data frame and the first row in data.

test_row = test_row2.loc[374, :]
data_row = test_drop.loc[0, :]
similarity (test_row, data_row, weights)
```

```python
# III Test the kNN Method 

#Apply the kNN method with k=5 on each of the test rows.

# 1. In data, add a new column named 'sim1' 
#that shows the similarity between each data row with the first test row.
#Display the first 5 rows of data.

test_row = test_row2.loc[374, :]
data_row = test_drop
sim = similarity (test_row, data_row, weights)

test_drop['sim1'] = sim
```
